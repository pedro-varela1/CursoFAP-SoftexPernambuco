# Atividade com Código 4 - FAP

## Instrução
Desenvolva um programa que deve ler um arquivo csv intitulado “notas_alunos.csv”. O arquivo deve ter a seguinte estrutura:

- aluno: Nome do aluno;
- nota_1: Primeira nota;
- nota_2: Segunda nota;
- faltas: Número de faltas;

O programa lerá esse arquivo e criará duas colunas. A primeira coluna será “media”, que terá a média das duas notas do aluno. A segunda será “situacao”, com os valores: APROVADO ou REPROVADO.

O aluno que tiver mais de 5 faltas ou possuir média menor que sete, será reprovado. O programa deverá salvar esse novo dataframe com o nome “alunos_situacao.csv”.

Por fim, o programa deverá mostrar na tela:
- o maior número de faltas;
- a média geral das notas dos alunos;
- e a maior média.

Veja em anexo um exemplo do arquivo “notas_alunos.csv”.

---

## Resolução

A resolução foi feita em Pyhton e está exposta abaixo, passo a passo.

Primeiramente, vamos criar um dictionary contendo os dados do arquivo .csv que será realizado a leitura:

````python
notas_alunos = {
    'aluno': 'Pedro Varela',
    'nota_1': 7,
    'nota_2': 10,
    'faltas': 5
}
````

Agora, vamos converter esse dict em um arquivo .csv (é importante ressaltar que essa etapa não era solicitada na questão, ao invés disso, era solicitado para fazer apenas a manipulação do arquivo _notas_alunos.csv_ e depois fazer o código processava esse arquivo anteriormente manipulado):

````python
import csv      #Biblioteca utilizada

with open('notas_alunos.csv', mode='w', newline='') as csvfile:
  # Vamos escrever um arquivo .csv com as keys em uma linha e os values em outra.
  w = csv.writer(csvfile)

  # Writing the header (keys)
  w.writerow(notas_alunos.keys())

  # Writing the values as a single row
  w.writerow(notas_alunos.values())
````

Arquivo criado, show de bola! Vamos transformar o arquivo em um _pandas dataframe_ para podermos dar uma olhada e manipulá-lo:

````python
import pandas as pd     # Vamos utilizar o pandas!

df_notas = pd.read_csv("/content/notas_alunos.csv")

df_notas # Uma checada é sempre bom
````

Observe o zero na esquerda, se trata do id daquele aluno no dataframe. Vamos mudar o id para a coluna _aluno_, ou seja, o id será o nome do aluno:

````python
df_notas = df_notas.set_index('aluno')
df_notas
````

Como solicitado, será criada a coluna da média em um novo dataframe:

````python
df_situacao = df_notas.copy()   # Dataframe novo copiado do antigo

df_situacao['media'] = (df_situacao['nota_1'] + df_situacao['nota_2'])/2    # Criacao da coluna da media

df_situacao
````

E a coluna da situção:

````python
df_situacao['situacao'] = df_situacao.apply(
    lambda x: 'APROVADO' if ((x['media']>=7) and (x['faltas']<=5)) else 'REPROVADO',
    axis=1
    )   # Condições estabelecidas para aprovação e reprovação

df_situacao
````

Vamos transformar esse dataframe novamente em um arquivo .csv:

````python
df_situacao.to_csv(path_or_buf='alunos_situacao.csv')
````
