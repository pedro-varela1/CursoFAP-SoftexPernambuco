# Atividade Insertion Sort - FAP

## Instrução
Faça um algoritmo de ordenação utilizando o método insertion sort.
Crie um método que execute as seguintes operações:

- Tamanho do vetor: 30;
- Utilize números ímpares;
- Opere em ordem crescente.

---

## Algoritmo

_Insertion sort_ é um algoritmo de classificação simples que funciona de modo semelhante a como você classifica as cartas de baralho em suas mãos. O _array_ é virtualmente dividido em uma parte classificada e outra não classificada. Os valores da parte não classificada são selecionados e colocados na posição correta na parte classificada.

Para classificar um _array_ de tamanho $N$ em ordem crescente, itere sobre o array e compare o elemento atual (_key_) com seu predecessor, se o elemento _key_ for menor que seu predecessor, compare-o com os elementos anteriores. Mova os elementos maiores uma posição para cima para liberar espaço para o elemento trocado.

## Código

Vamos fazer uma função implementando o algoritmo estudado, operando em ordem crescente:

```python
def insertionSort(vec):

  # Atravesse 1 até o tamanho do vetor
  for i in range(1, len(vec)):

    key = vec[i]

    # Mova elementos de vec[0..i-1], que são maiores que key, para uma posição à frente de sua posição atual
    j = i-1
    while j >= 0 and key < vec[j] :
            vec[j + 1] = vec[j]
            j -= 1
    vec[j + 1] = key

  return vec
```

Vamos testar a função _insertionSort_:

```python
# Criação de vetor de tamanho 30 com números ímpares, como solicitado
vetor = [9, 11, 13, 3, 1, 77, 5, 7, 33, 121, 67, 17, 19, 29, 27, 11, 13, 91, 97, 3, 1, 1, 7, 95, 47, 33, 85, 63, 9, 37]

len(vetor)    # Checar tamanho do vetor
```

```python
# Ordenar o vetor com a função criada
vetor_sort = insertionSort(vetor)

print(vetor_sort)
```

Vemos que a função _insertionSort_ ordenou o _array_ com êxito, utilizando a lógica mostrada no algoritmo.
