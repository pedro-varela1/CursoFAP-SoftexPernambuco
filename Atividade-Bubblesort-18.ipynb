# Atividade Bubble Sort - FAP

## Instrução
Construa um algoritmo de ordenação, utilizando o método bubble sort estudado. (Lembre-se que se trata de uma série de instruções que devem ser seguidas passo a passo).
Para isso, você deve criar um método em que o tamanho do vetor seja dez e deve estar em ordem crescente.

O vetor deverá:
- comparar seus elementos dois a dois adjacentes;
- se os elementos não estiverem em ordem, então ordenar;
- senão, avançar para o próximo par;
- repetir a operação até que nenhuma troca possa ser feita no vetor inteiro.

---

## Algoritmo

A resolução foi feita em Pyhton e está exposta passo a passo. Antes de analisá-lo, vamos a série de passos feitas no Bubble Sort tendo como exemplo o array [6, 0, 3, 5].

1. Primeiramente, colocamos o maior elemento na posição correta, ou seja, no final da lista.
image.png

2. O segundo passo é colocar o segundo maior elemento na posição correta.
image.png

3. Por fim, basta colocar os dois últimos elementos na posição correta.
image.png

Observe que a quantidade de passos depende do tamanho do array cuja ordenação será feita, seguindo o padrão $n+1$ passos - com $n$ sendo o tamanho do array.

---

## Código

Vamos fazer uma função implementando o algoritmo estudado:

```python
def bubbleSort(vec):
  n = len(vec)    # n armazena o tamanho do vetor

  for i in range(n):
    swapped = False   # a variável swapped é true se houver alguma troca de posição durante a iteração principal

    for j in range(0, n-i-1):   # Percorre os primeiros elementos com relação ao valor de i, o -1 vem em decorrência de i começar em 0
      if vec[j] > vec[j+1]:   # Condição para se o valor não estiver na posição correta
        vec[j], vec[j+1] = vec[j+1], vec[j]
        swapped = True
    if (swapped == False):    # Se não houver troca de posição, vetor já está ordenado, e saímos da função
        break

  return vec
```

Vamos testar a função _bubbleSort_:

```python
vetor = [6, 0, 3, 5]    # Vetor usado na explicação do algoritmo

vetor_sort = bubbleSort(vetor)    # Ordenar o vetor com a função criada

vetor_sort
```

Vemos que a função _bubbleSort_ ordenou o _array_ com êxito, utilizando a lógica mostrada no algoritmo.
